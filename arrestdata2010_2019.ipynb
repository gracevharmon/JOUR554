```python
arrest_data = pd.read_csv('https://raw.githubusercontent.com/gracevharmon/JOUR554/7a274698d44edb4f9bfc870f573859c03afcbc3b/Arrest_Data_from_2010_to_2019.csv')
```

```python
arrest_data.columns = arrest_data.columns.str.lower().str.replace(' ' ,'_') 
```

```python
arrest_data.columns
```

```python
df = arrest_data[['age', 'sex_code', 'descent_code',
       'charge_group_code', 'charge_group_description','charge', 'charge_description','address','lat', 'lon', 'location']]
```

```python
ages = df.groupby(['charge_group_description']).agg({'age':'mean'}).round(0).reset_index()
```
